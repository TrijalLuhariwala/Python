## Encoding

```python
df.drop(["unique_id"],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df["age_bracket"]=le.fit_transform(df["age_bracket"])
df["qualification"]=le.fit_transform(df["qualification"])
df["race"]=le.fit_transform(df["race"])
df["sex"]=le.fit_transform(df["sex"])
df["income_level"]=le.fit_transform(df["income_level"])
df["marital_status"]=le.fit_transform(df["marital_status"])
df["housing_status"]=le.fit_transform(df["housing_status"])
df["employment"]=le.fit_transform(df["employment"])
df["census_msa"]=le.fit_transform(df["census_msa"])
```

```python
df.dtypes
```

```python
df.sample(15).style.set_properties(
    **{
        'background-color':'yellow',
        'color':'blue',
        'border-color':'magenta'
    })
```

## Spliting Data

```python
X=df.drop(columns="h1n1_vaccine",axis=1)
y=df["h1n1_vaccine"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

## Classification Models

### 1.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression(solver="lbfgs")
model1.fit(X_train,y_train)
model1
```

```python
model1.score(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

#### Logistic Regression is performing good with both test and train.

### 2.KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model2=KNeighborsClassifier(n_neighbors=30)
model2.fit(X_train,y_train)
model2
```

```python
model2.score(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

#### KNN Classifier is performing good with both test and train.

### 3.Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model3=DecisionTreeClassifier(max_depth=7,max_features=11)
model3.fit(X_train,y_train)
model3
```

```python
model3.score(X_train,y_train)
```

```python
model3.score(X_test,y_test)
```

#### Decision Tree Classifier is performing good with both test and train.

### 4.Naive Bayes Theorum

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
model4=CategoricalNB()
model4.fit(X_train,y_train)
model4
```

```python
model4.score(X_train,y_train)
```

```python
model4.score(X_test,y_test)
```

#### Naive Bayes Theorum is performing averagely with both test and train.

### 5.Bagging Classifier

```python
from sklearn.ensemble import BaggingClassifier
```

```python
model5=BaggingClassifier(n_estimators=40,max_features=7) 
model5.fit(X_train, y_train)
y_predict5=model5.predict(X_test)
model5
```

```python
model5.score(X_train,y_train)
```

```python
model5.score(X_test,y_test)
```

#### Bagging Classifier is performing averagely with both test and train.

### 6.Ada Boost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model6=AdaBoostClassifier(n_estimators=40)
model6.fit(X_train,y_train)
y_predict6=model6.predict(X_test)
model6
```

```python
model6.score(X_train,y_train)
```

```python
model6.score(X_test,y_test)
```

#### Ada Boost Classifier is performing good with both test and train.

### 7.Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model7=GradientBoostingClassifier()
model7.fit(X_train,y_train)
y_predict7=model7.predict(X_test)
model7
```

```python
model7.score(X_train,y_train)
```

```python
model7.score(X_test,y_test)
```

#### Gradient Boosting Classifier is performing good with both test and train.

### 8.Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model8=RandomForestClassifier(n_estimators=60,max_depth=7)
model8.fit(X_train,y_train)
y_predict8=model8.predict(X_test)
model8
```

```python
model8.score(X_train,y_train)
```

```python
model8.score(X_test,y_test)
```

#### Random Forest Classifier is performing good with both test and train.

### Confusion Matrix

```python
from sklearn import metrics
```

```python
cm=metrics.confusion_matrix(y_test,y_predict7,labels=[0,1])

df_cm=pd.DataFrame(cm,index=[i for i in["No","Yes"]],
                  columns=[i for i in["No","Yes"]])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True,fmt='g')
```
